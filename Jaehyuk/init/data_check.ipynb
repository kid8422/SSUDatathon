```python
import pandas as pd
import re

# 4자리 연도 추출 및 2025년 이전의 가장 최근 연도 선택 함수
def extract_publication_year(text):
    # 텍스트에서 모든 4자리 숫자 추출
    if pd.isna(text):
        return None
    
    years = re.findall(r'\b\d{4}\b', str(text))
    
    # 찾은 연도가 없으면 None 반환
    if not years:
        return None
    
    # 4자리 연도를 정수형으로 변환하고 2025년보다 작은 값만 필터링
    valid_years = [int(year) for year in years if int(year) < 2025]
    
    # 유효한 연도가 없으면 None 반환
    if not valid_years:
        return None
    
    # 가장 마지막에 등장한 유효한 연도를 선택하여 반환 (int 타입)
    return int(max(valid_years))

# 예시 데이터프레임
df = pd.read_csv('/mnt/c/Users/allo1/OneDrive/바탕 화면/GitHub/SSUDatathon/Data/단행본(도서)정보.txt', delimiter=',', encoding='euc-KR')

# '출판년도' 열에 대해 함수 적용
df['출판년도'] = df['출판년도'].apply(extract_publication_year)

# 결과 출력
print(df[['출판년도']].head())
```

```python
import pandas as pd
import re

# 텍스트 파일을 읽어옵니다
df = pd.read_csv('/mnt/c/Users/allo1/OneDrive/바탕 화면/GitHub/SSUDatathon/Data/단행본(도서)정보.txt', delimiter=',', encoding='euc-KR')

# 4자리 연도가 존재하는지 확인하는 함수
def has_four_digit_year(text):
    if pd.isna(text):
        return False
    
    # 텍스트에서 4자리 숫자 찾기
    years = re.findall(r'\b\d{4}\b', str(text))
    
    # 4자리 숫자가 하나라도 있으면 True, 없으면 False
    return bool(years)

# '출판년도' 열에서 4자리 연도가 없는 행 번호 찾기
no_years_rows = df[~df['출판년도'].apply(has_four_digit_year)].index

# 결과 출력
print("출판년도에 4자리 정수가 없는 행 번호:", no_years_rows.tolist())
```
