```python
import pandas as pd

df = pd.read_csv('/mnt/c/Users/allo1/OneDrive/바탕 화면/GitHub/SSUDatathon/Data/단행본(도서)정보.txt', delimiter=',', encoding='euc-KR')
df
```

```python
# 각 열별로 가장 긴 문자열의 길이 계산
max_lengths = df.astype(str).applymap(len).max()

# 결과 출력
print(max_lengths)
```

```python
# 각 행의 문자열 길이를 기준으로 정렬하고 상위 50개 선택
top_50_longest_publication_year = df.assign(출판년도길이=df['출판년도'].str.len()) \
                                    .sort_values(by=['출판년도길이'], ascending=False) \
                                    .head(50)['출판년도']

# 결과 출력
print(top_50_longest_publication_year)
```
