```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

# .env 파일 로드
load_dotenv()

# 환경 변수에서 DB 정보 가져오기
DB_HOST = os.getenv("DB_IP")
DB_PORT = os.getenv("DB_PORT")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
DB_NAME = "hardcoding"
TABLE_NAME = "year_month_count_detail"

# SQLAlchemy 연결 문자열 생성
DATABASE_URL = f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

# 데이터베이스 연결
engine = create_engine(DATABASE_URL)

# 특정 테이블을 pandas 데이터프레임으로 변환
df = pd.read_sql(f"SELECT * FROM {TABLE_NAME}", con=engine)

# 데이터 확인
print(df.head())

# 연결 종료
engine.dispose()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"  # 실제 존재하는지 확인

# 폰트 로드
font_prop = fm.FontProperties(fname=font_path)

# matplotlib 기본 폰트 설정
plt.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# **연도별 합계 계산 (month 제거)**
df_grouped = df.drop(columns=["month"]).groupby("year").sum()

# 그래프 설정
plt.figure(figsize=(12, 6))
colors = ['green', 'red', 'gray', 'orange', 'brown', 'cyan', 'gold', 'lightgreen', 'blue', 'purple']

# 꺾은선 그래프 그리기
for i, column in enumerate(df_grouped.columns):
    plt.plot(df_grouped.index, df_grouped[column], marker='o', linestyle='-', label=column, color=colors[i])

# X축을 정수형으로 변환
plt.xticks(df_grouped.index, df_grouped.index.astype(int))

# 그래프 제목 및 라벨 설정
plt.title("연도, 대분류별 대출 권수 분포", fontsize=14)
plt.xlabel("연도", fontsize=12)
plt.ylabel("도서 수", fontsize=12)
plt.legend(title="대분류", loc="upper left", bbox_to_anchor=(1,1))
plt.grid(True, linestyle="--", alpha=0.6)

# 그래프 출력
plt.show()
```

```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

# .env 파일 로드
load_dotenv()

# 환경 변수에서 DB 정보 가져오기
DB_HOST = os.getenv("DB_IP")
DB_PORT = os.getenv("DB_PORT")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
DB_NAME = "hardcoding2"
TABLE_NAME = "ISBN_rent_count"

# SQLAlchemy 연결 문자열 생성
DATABASE_URL = f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

# 데이터베이스 연결
engine = create_engine(DATABASE_URL)

# 특정 테이블을 pandas 데이터프레임으로 변환
df = pd.read_sql(f"SELECT * FROM {TABLE_NAME}", con=engine)

# 데이터 확인
print(df.head())

# 연결 종료
engine.dispose()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ISBN 제외한 데이터 (연도별 데이터만 사용)
df_without_isbn = df.drop(columns=['ISBN'])

# ID_count 컬럼을 기반으로 크기와 색깔 결정
sizes = df['ID_count'] * 10  # 점의 크기 (ID_count 값에 따라 확대)
colors = np.log(df['ID_count'] + 1)  # ID_count 값에 대한 색상 (log transform for better spread)

# 연도별로 데이터를 나열
years = list(map(str, range(2004, 2025)))

# 산점도 그리기
plt.figure(figsize=(10, 6))
for year in years:
    # 각 연도에 대해 데이터 값 추출
    values = df_without_isbn[year]
    
    # 해당 연도의 산점도를 그리기
    scatter = plt.scatter([year] * len(values), values, s=sizes, c=colors, cmap='plasma', alpha=0.6)

# 색상 바 추가 (ID_count에 대한 색상 맵)
plt.colorbar(scatter, label='log(ID_count + 1)')

# 그래프 레이블 및 제목 설정
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Scatter Plot by Year with ID_count based size and color')
plt.xticks(rotation=45)

# 그래프 출력
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ISBN 제외한 데이터 (연도별 데이터만 사용)
df_without_isbn = df.drop(columns=['ISBN'])

# ID_count 컬럼을 기반으로 크기와 색깔 결정
sizes = df['ID_count'] * 10  # 점의 크기 (ID_count 값에 따라 확대)
colors = np.log(df['ID_count'] + 1)  # ID_count 값에 대한 색상 (log transform for better spread)

# 연도별로 데이터를 나열 (문자형으로 변환)
years = list(map(str, range(2004, 2025)))

# 산점도 그리기
plt.figure(figsize=(10, 6))
for year in years:
    # 각 연도에 대해 데이터 값 추출
    values = df_without_isbn[year]
    
    # 해당 연도의 산점도를 그리기
    scatter = plt.scatter([year] * len(values), values, s=sizes, c=colors, cmap='plasma', alpha=0.6)

# 색상 바 추가 (ID_count에 대한 색상 맵)
plt.colorbar(scatter, label='log(ID_count + 1)')

# y축 범위 조정: 전체 데이터의 최소/최대 값을 계산하고, 10% 여유 공간 추가
y_min = df_without_isbn[years].min().min()
y_max = df_without_isbn[years].max().max()
padding = (y_max - y_min) * 0.1  # 10% 여유
plt.ylim(y_min - padding, y_max + padding)

# 그래프 레이블 및 제목 설정
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Scatter Plot by Year with ID_count based size and color')
plt.xticks(rotation=45)

# 그래프 출력
plt.tight_layout()
plt.show()
```
